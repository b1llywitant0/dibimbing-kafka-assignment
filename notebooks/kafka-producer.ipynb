```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaProducer
from faker import Faker
from time import sleep
```

```python
# Using the variables in .env file
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
# Testing one of the value inside .env
kafka_host = os.getenv('KAFKA_HOST')
```

```python
# Creating Producers for Kafka
producer = KafkaProducer(bootstrap_servers=f'{kafka_host}:9092')
```

```python
# Using Faker
_instance = Faker()
global faker
faker = Faker()
```

```python
class DataGenerator(object):
    @staticmethod
    def get_data():
        return [
                uuid.uuid4().__str__(),
                faker.name(),
                faker.random_element(elements=('Meta', 'Google', 'TikTok', 'Programmatic')),
                faker.random_element(elements=('DKI Jakarta', 'Jawa Barat', 'Jawa Timur', 'Jawa Tengah', 'Banten')),
                faker.random_int(min=18, max=45),
                uuid.uuid4().__str__(),
                faker.random_int(min=100000, max=10000000),
                faker.unix_time()
        ]

for i in range(1,400):
    columns =  ["lead_id", "lead_name", "ads_channel", "province", "age", "transaction_id", "purchase_value", "ts"]
    data_list = DataGenerator.get_data()
    json_data = dict(
        zip(columns,data_list)
    )
    _payload = json.dumps(json_data).encode("utf-8")
    
    
    
    response = producer.send(topic='kafka-assignment', value=_payload)
    print(json_data['lead_id'],response.get())
    sleep(5)
```
